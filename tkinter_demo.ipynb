```python
# Button + messagebox + addition function
import tkinter
from tkinter import messagebox
win=tkinter.Tk()
win.title('GUI')

def add():
    #print('Button clicked')
    a=5
    b=6
    sum=float(a)+float(b)
    messagebox.showinfo('addition',sum)

b=tkinter.Button(win,text='button demo of addition',bg='red',fg='white',command=add)
b.grid(column=0,row=0)

win.mainloop()
```

```python
#  Entry widget + addition substraction + history of all calculation
import tkinter
win=tkinter.Tk()
win.geometry('300x300')
win.title(' Entry widget ')

#text.pack()
labelval1=tkinter.Label(win,text='number 1 :')
labelval1.grid(column=0,row=0)
text1=tkinter.Entry(win)
text1.grid(column=1,row=0)

labelval2=tkinter.Label(win,text='number 2 :')
labelval2.grid(column=0,row=1)
text2=tkinter.Entry(win)
text2.grid(column=1,row=1)

#------------------------------- addition--------------------------------------
def add():
    #print('Button clicked')
    num1=text1.get()
    num2=text2.get()
    sum=float(num1)+float(num2)
    #messagebox.showinfo('addition',sum)
    labelres=tkinter.Label(win,text='Result :')
    labelres.configure(text=sum)
    labelres.grid(column=0)
    
badd=tkinter.Button(win,text='+',bg='red',fg='white',command=add)
badd.grid(column=0,row=3)
#-------------------------------- substraction -------------------------------------
def sub():
    #print('Button clicked')
    num1=text1.get()
    num2=text2.get()
    minus=float(num1)-float(num2)
    #sub=float(num1) - float(num2)
    labelres=tkinter.Label(win,text='Result :')
    labelres.configure(text=minus)
    labelres.grid(column=1)
    
bsub=tkinter.Button(win,text='-',bg='red',fg='white',command=sub)
bsub.grid(column=1,row=3)

win.mainloop()
```

```python
# ComboBox : drop down menu with text
# ex : online persoal info filling form
import tkinter
from tkinter import ttk
win=tkinter.Tk()
win.title(' ComboBox ')
win.geometry('300x300')

labelcity=tkinter.Label(win,text='Select Your City :')
labelcity.grid()

cbox=ttk.Combobox(win)
cbox['values']=('Select','Pune','Nashik','Parbhani','Aurangabad')
cbox.current(0)
cbox.grid()

win.mainloop()
```

```python
# CheckButton
import tkinter
win=tkinter.Tk()
win.title(' CheckButton ')
win.geometry('300x300')

check_state1=tkinter.BooleanVar()
check_state1.set(False)

check_state2=tkinter.BooleanVar()
check_state2.set(False)

check_state3=tkinter.BooleanVar()
check_state3.set(False)

check_state4=tkinter.BooleanVar()
check_state4.set(False)

check_state5=tkinter.BooleanVar()
check_state5.set(False)

check_state6=tkinter.BooleanVar()
check_state6.set(False)

labelsub=tkinter.Label(win,text='Select Your Subjects :')
labelsub.grid(column=0, row=0)

cb1=tkinter.Checkbutton(win,text='Python',var=check_state1)
cb1.grid(column=0, row=1)

cb2=tkinter.Checkbutton(win,text='DS',var=check_state2)
cb2.grid(column=1, row=1)

cb3=tkinter.Checkbutton(win,text='Java',var=check_state3)
cb3.grid(column=0, row=2)

cb4=tkinter.Checkbutton(win,text='OS',var=check_state4)
cb4.grid(column=1, row=2)

cb5=tkinter.Checkbutton(win,text='C++',var=check_state5)
cb5.grid(column=0, row=3)

cb6=tkinter.Checkbutton(win,text='C',var=check_state6)
cb6.grid(column=1, row=3)


win.mainloop()
```

```python
# Message Box
import tkinter
from tkinter import messagebox
win=tkinter.Tk()
win.title(' Message Box ')
win.geometry('300x300')

def button_clicked():
    messagebox.showinfo('Message Box','This is Message Box')

b=tkinter.Button(win,text='Click Here', bg='Red', fg='Yellow', command=button_clicked)
b.grid()

win.mainloop()
```

```python
# Menu Bar
import tkinter
win=tkinter.Tk()
win.title(' Menu Bar ')
win.geometry('300x300')

menubar=tkinter.Menu(win)


filemenu=tkinter.Menu(menubar)

filemenu.add_command(label='New')
filemenu.add_command(label='Open')
filemenu.add_command(label='Save')
filemenu.add_command(label='Delete')
filemenu.add_command(label='Quit')

menubar.add_cascade(label='File', menu=filemenu)

menubar.add_command(label='Help')
menubar.add_command(label='References')
#menubar.add_command(label='References')


win.config(menu=menubar)

win.mainloop()
```
