```python
#  Entry widget + addition substraction + history of all calculation
import tkinter
win=tkinter.Tk()
win.geometry('300x300')
win.title(' Entry widget ')

labelval1=tkinter.Label(win,text='number 1 :')
labelval1.grid(column=0,row=0)
text1=tkinter.Entry(win)
text1.grid(column=1,row=0)

labelval2=tkinter.Label(win,text='number 2 :')
labelval2.grid(column=0,row=1)
text2=tkinter.Entry(win)
text2.grid(column=1,row=1)

#res=0.0


#------------------------------- addition--------------------------------------
def add():
    num1=float(text1.get())
    num2=float(text2.get())
    #global res
    res=float(num1)+float(num2)
    formatted_float = "{:.2f}".format(res)
    res=formatted_float
    
    lsum = tkinter.Label(win, text = 'addition')
    lsum.grid(row=7, column=1)
    lsum["text"] = " SUM :  " + str(res)
    
badd=tkinter.Button(win,text=' + ',bg='red',fg='white',command=add)
badd.grid(column=0,row=3)

#-------------------------------- subtraction -------------------------------------
def sub():
    #print('Button clicked')
    num1=text1.get()
    num2=text2.get()
    #global res
    res=float(num1)-float(num2)
    formatted_float = "{:.2f}".format(res)
    res=formatted_float
    
    lsum = tkinter.Label(win, text = 'subtraction')
    lsum.grid(row=7, column=1)
    lsum["text"] = " SUB :  " + str(res)
    
bsub=tkinter.Button(win,text=' - ',bg='red',fg='white',command=sub)
bsub.grid(column=1,row=3)
#bsub.grid()

#------------------------------- Multiplication --------------------------------------
def mul():
    num1=text1.get()
    num2=text2.get()
    #global res
    res=float(num1) * float(num2)
    formatted_float = "{:.2f}".format(res)
    res=formatted_float
    
    lsum = tkinter.Label(win, text = 'multiplication')
    lsum.grid(row=7, column=1)
    lsum["text"] = " MUL :  " + str(res)
    
bmul=tkinter.Button(win,text=' * ',bg='red',fg='white',command=mul)
bmul.grid(column=0,row=5)

#-------------------------------- Division -------------------------------------
def div():
    num1=text1.get()
    num2=text2.get()
    #global res
    res=float(num1)/float(num2)
    formatted_float = "{:.2f}".format(res)
    res=formatted_float
    
    lsum = tkinter.Label(win, text = 'division')
    lsum.grid(row=7, column=1)
    lsum["text"] = " DIV :  " + str(res)
    
bdiv=tkinter.Button(win,text=' / ',bg='red',fg='white',command=div)
bdiv.grid(column=1,row=5)
#bdiv.grid()


win.mainloop()
```
